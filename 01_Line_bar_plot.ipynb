## Adding tittle

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load data_set
phool= sns.load_dataset("iris")
phool

# draw a line plot and it always plot b/t two numerical number
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Pholo ka plot")
plt.show()
```

## Adding limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load data_set
phool= sns.load_dataset("iris")
phool

# draw a line plot and it always plot b/t two numerical number
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Pholo ka plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt



# load data_set
phool= sns.load_dataset("iris")
phool


# draw a line plot and it always plot b/t two numerical number
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Pholo ka plot")
sns.set_style("black")
sns.set_style(style=None, rc=None)

plt.show()
```

## size of figers

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt



# load data_set
phool= sns.load_dataset("iris")
phool
sns.set_style("dark")
plt.figure(figsize=(6,8))

# draw a line plot and it always plot b/t two numerical number
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Pholo ka plot")


plt.show()
```
